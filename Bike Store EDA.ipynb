### `Customers_Age` üzrə datanın ədədi ortası nədir?

```python
# your code goes here
print('Mean of Customer Age:',sales['Customer_Age'].mean())
```

Bu datanı (`Customer_Age`) <b>density (KDE)</b> və <b>box plot</b> olaraq ifadə edin:

```python
# your code goes here
#Boxplot
sales['Customer_Age'].plot(kind = 'box',figsize = (14,6),vert = False,fontsize = 12)
```

```python
sales['Customer_Age'].plot(kind = 'kde',figsize = (14,6))
```

### `Order_Quantity` üzrə datanın ədədi ortası nədir??

```python
# your code goes here
print('Mean of Order Quantity:',sales['Order_Quantity'].mean())
```

`Order_Quantity` üzrə datanın <b>histogram</b> və <b>box plot</b>-unu göstərin:

```python
# your code goes here
plt.figure(figsize =(14,6),dpi = 300)
sns.histplot(data = sales,x = 'Order_Quantity');
```

```python
sales['Order_Quantity'].plot(kind='box',figsize = (14,6),vert = False)
```

### Illər üzrə satış miqdarı nə qədərdir?

```python
# your code goes here
sales['Year'].value_counts()
```

Bunu <b>pie plot</b> şəklində göstərin (illər üzrə faizlərə bölməklə):

```python
# your code goes here
sales['Year'].value_counts().plot(kind = 'pie',figsize = (20,10),autopct='%1.1f%%')
plt.title('Sales per year',fontsize = 15)
```

### Aylar üzrə satış miqdarı nə qədərdir?

```python
# your code goes here
sales['Month'].value_counts()
```

Bunu <b>bar plot</b> şəklində göstərin:

```python
# your code goes here
sales['Month'].value_counts().plot(kind = 'bar',figsize = (14,6))
plt.title('Sales per month',fontsize = 17)
plt.legend(loc = 'best')
```

### Ən yuxarı **satış sayı**na sahib ölkə hansıdır ?

```python
# your code goes here
sales['Country'].value_counts()
```

Hər ölkə üzrə satışları <b>bar plot</b> şəklində göstərin:

```python
# your code goes here
sales['Country'].value_counts().plot(kind = 'bar',figsize = (14,6))
plt.title('Sales in each country',fontsize = 17)
plt.legend(loc = 'best')
```

### Satılmış hər bir məhsulun göstərildiyi list yaradın

```python
# your code goes here
sales['Product'].value_counts()
```

10 ən çox satılmış məhsulun satış miqdarını göstərən **bar plot** qurun (best sellers):

```python
# your code goes here
sales['Product'].value_counts().nlargest(10).plot(kind = 'bar',figsize = (14,6))
plt.title('Best seller 10 products',fontsize = 17)
plt.xlabel('Products')
plt.ylabel('Quantity of products')
plt.legend(loc='best')
```

### `Unit_Cost` və `Unit_Price` arasında necə bir əlaqə görə bilirsiniz?

Bunu plot-larla (uyğun gördüyünüz tip) bildirərək izah edin

```python
# your code goes here
sales.plot(kind = 'scatter',x = 'Unit_Cost',y = 'Unit_Price',figsize = (14,6))
plt.title('Relations between Unit Cost and Unit Price',fontsize = 17)
plt.xlabel('Unit Cost',fontsize = 14)
plt.ylabel('Unit Price',fontsize = 14)
plt.show()
```

### `Order_Quantity` and `Profit` arasında necə bir əlaqə görə bilirsiniz?
